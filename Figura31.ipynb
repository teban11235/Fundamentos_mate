```python
import matplotlib.pyplot as plt

# Coordenadas de los puntos A y B
punto_A = (-2, 0)
punto_B = (4, 0)

# Crear la recta (eje x)
x = [-3, 5]  # Rango de valores para x
y = [0, 0]   # Constante en 0 porque est√° sobre el eje x

# Crear la gr√°fica
plt.figure(figsize=(8, 4))

# Graficar la recta
plt.plot(x, y, color='pink', linewidth=2, label='Recta sobre el eje X')

# Graficar los puntos A y B
plt.scatter(punto_A[0], punto_A[1], color='blue', s=100, label='Punto A (-2, 0)')
plt.scatter(punto_B[0], punto_B[1], color='blue', s=100, label='Punto B (4, 0)')

# Agregar etiquetas a los puntos
plt.text(punto_A[0] - 0.5, punto_A[1] + 0.2, 'A', color='blue', fontsize=12)
plt.text(punto_B[0] + 0.2, punto_B[1] + 0.2, 'B', color='blue', fontsize=12)

# Configuraci√≥n de la gr√°fica
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje X
plt.axvline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje Y
#plt.title('Recta y Puntos A y B sobre el Eje X')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(range(-3, 6))  # Etiquetas de los valores en x
plt.yticks(range(-1, 2))  # Etiquetas de los valores en y
plt.grid(alpha=0.4)
plt.legend()

# Mostrar la gr√°fica
plt.show()
```
