```python
import matplotlib.pyplot as plt
import numpy as np

# Par√°metros para la primera elipse (centrada en el origen, eje mayor horizontal)
a1, b1 = 5, 3  # Semieje mayor y menor
c1 = np.sqrt(a1**2 - b1**2)  # Distancia focal
focos1 = [(-c1, 0), (c1, 0)]  # Focos de la primera elipse
centro1 = (0, 0)  # Centro de la primera elipse

# Par√°metros para la segunda elipse (centrada en (3, -2), eje mayor vertical)
a2, b2 = 3, 5  # Semieje menor (horizontal) y mayor (vertical)
c2 = np.sqrt(b2**2 - a2**2)  # Distancia focal
centro2 = (3, -2)  # Centro de la segunda elipse
focos2 = [(centro2[0], centro2[1] - c2), (centro2[0], centro2[1] + c2)]  # Focos de la segunda elipse

# Crear valores de theta para graficar las elipses
theta = np.linspace(0, 2 * np.pi, 500)

# Coordenadas de la primera elipse (centrada en el origen, horizontal)
x1 = a1 * np.cos(theta)
y1 = b1 * np.sin(theta)

# Coordenadas de la segunda elipse (centrada en (3, -2), vertical)
x2 = centro2[0] + a2 * np.cos(theta)
y2 = centro2[1] + b2 * np.sin(theta)

# Crear la gr√°fica
plt.figure(figsize=(10, 8))

# Graficar la primera elipse
plt.plot(x1, y1, label='Elipse 1: Centro (0, 0), eje mayor horizontal', color='blue')
plt.scatter(*zip(*focos1), color='red', label='Focos Elipse 1', zorder=5)

# Graficar el centro de la primera elipse
plt.scatter(centro1[0], centro1[1], color='purple', label=f'Centro Elipse 1 {centro1}', zorder=5)
plt.text(centro1[0] + 0.2, centro1[1] + 0.2, f'{centro1}', color='purple', fontsize=10)

# Graficar la segunda elipse
plt.plot(x2, y2, label='Elipse 2: Centro (3, -2), eje mayor vertical', color='green')
plt.scatter(*zip(*focos2), color='orange', label='Focos Elipse 2', zorder=5)

# Graficar el centro de la segunda elipse
plt.scatter(centro2[0], centro2[1], color='brown', label=f'Centro Elipse 2 {centro2}', zorder=5)
plt.text(centro2[0] + 0.2, centro2[1] + 0.2, f'{centro2}', color='brown', fontsize=10)

# Configuraci√≥n de la gr√°fica
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje X
plt.axvline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje Y
#plt.title('Elipses con eje mayor horizontal y vertical')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')  # Escalas iguales para los ejes
plt.grid(alpha=0.4)
plt.legend()

# Mostrar la gr√°fica
plt.show()
```
