```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para la función y = 2x - 1
x = np.array([-3.0, -2.0, -1.5, -1.0, -0.5, 0, 0.5, 1.0, 1.5, 2.0, 3.0])
y = 2 * x - 1

# Crear la gráfica solo con puntos
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='green', label='y = 2x - 1')

# Configuración del gráfico
plt.axhline(0, color='black', linewidth=0.8, linestyle='--', alpha=0.6)
plt.axvline(0, color='black', linewidth=0.8, linestyle='--', alpha=0.6)
plt.xlabel('x')
plt.ylabel('y = 2x - 1')
#plt.title('Gráfica de la función y = 2x - 1 (solo puntos)')
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```
