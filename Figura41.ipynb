```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos para las tres pendientes
x = np.linspace(-10, 10, 100)  # Valores de x comunes

# Ecuaciones de las rectas
y_positiva = 2 * x + 1  # Pendiente positiva
y_negativa = -2 * x - 1  # Pendiente negativa
y_nula = np.zeros_like(x)  # Pendiente nula (y = 0)

# Crear la gr√°fica
plt.figure(figsize=(10, 6))

# Graficar las rectas
plt.plot(x, y_positiva, label='Pendiente positiva (y = 2x + 1)', color='blue')
plt.plot(x, y_negativa, label='Pendiente negativa (y = -2x - 1)', color='red')
plt.plot(x, y_nula, label='Pendiente nula (y = 0)', color='green')

# Configuraci√≥n de la gr√°fica
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje X
plt.axvline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje Y
plt.xlabel('x')
plt.ylabel('y')
#plt.title('Gr√°fica de tres tipos de pendientes')
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
