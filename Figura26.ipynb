```python
import matplotlib.pyplot as plt

# Coordenadas de los puntos
puntos = {
    "A": (2, 3),
    "B": (-1, 1),
    "C": (1, -1),
    "D": (-2, -2),
}

# Crear la gr√°fica
plt.figure(figsize=(8, 8))

# Dibujar los puntos y a√±adir etiquetas
for etiqueta, (x, y) in puntos.items():
    plt.scatter(x, y, color='blue')  # Puntos
    plt.text(x + 0.1, y + 0.1, f'{etiqueta} = ({x}, {y})', fontsize=10, color='blue')  # Etiquetas

# Configuraci√≥n de la gr√°fica
plt.axhline(0, color='black', linewidth=0.8, linestyle='--', alpha=0.6)  # Eje X
plt.axvline(0, color='black', linewidth=0.8, linestyle='--', alpha=0.6)  # Eje Y
plt.grid(alpha=0.4)
plt.xlabel('x')
plt.ylabel('y')
#plt.title('Gr√°fica de los puntos A, B, C y D')
plt.xticks(range(-3, 4))
plt.yticks(range(-3, 5))
plt.xlim(-3, 3)
plt.ylim(-3, 4)
plt.show()
```
