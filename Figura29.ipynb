```python
import matplotlib.pyplot as plt
import numpy as np

# Datos extraídos de la tabla
x = np.array([-3.0, -2.0, -1.5, -1.0, -0.5, 0.5, 1.0, 1.5, 2.0, 3.0])
y = np.array([-1/3, -1/2, -2/3, -1, -2, 2, 1, 2/3, 1/2, 1/3])

# Crear la gráfica solo con puntos
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='y = 1/x')

# Configuración del gráfico
plt.axhline(0, color='black', linewidth=0.8, linestyle='--', alpha=0.6)
plt.axvline(0, color='black', linewidth=0.8, linestyle='--', alpha=0.6)
plt.xlabel('x')
plt.ylabel('y = 1/x')
#plt.title('Gráfica de la función y = 1/x (solo puntos)')
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```
