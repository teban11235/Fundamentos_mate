```python

import matplotlib.pyplot as plt
import numpy as np

# Coordenadas de los v√©rtices
A = (-1, -3)
B = (6, 1)
C = (2, -7)

# Funci√≥n para calcular la distancia entre dos puntos
def distancia(p1, p2):
    return np.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)

# Calcular las longitudes de los lados
lado_AB = distancia(A, B)
lado_BC = distancia(B, C)
lado_CA = distancia(C, A)

# Mostrar las longitudes de los lados
print(f"Longitud del lado AB: {lado_AB:.2f}")
print(f"Longitud del lado BC: {lado_BC:.2f}")
print(f"Longitud del lado CA: {lado_CA:.2f}")

# Graficar el tri√°ngulo
plt.figure(figsize=(8, 6))

# Trazar l√≠neas entrecortadas con colores distintos
plt.plot([A[0], B[0]], [A[1], B[1]], linestyle='--', color='blue', label='Lado AB')  # L√≠nea entre A y B
plt.plot([B[0], C[0]], [B[1], C[1]], linestyle='--', color='green', label='Lado BC')  # L√≠nea entre B y C
plt.plot([C[0], A[0]], [C[1], A[1]], linestyle='--', color='orange', label='Lado CA')  # L√≠nea entre C y A

# Etiquetar los v√©rtices con nombres y coordenadas
plt.scatter([A[0], B[0], C[0]], [A[1], B[1], C[1]], color='red', zorder=5)  # Puntos de los v√©rtices
plt.text(A[0], A[1], f'A{A}', fontsize=12, color='red', ha='right')
plt.text(B[0], B[1], f'B{B}', fontsize=12, color='red', ha='left')
plt.text(C[0], C[1], f'C{C}', fontsize=12, color='red', ha='left')

# Configuraci√≥n del gr√°fico
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)
plt.axvline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)
plt.grid(alpha=0.4)
plt.xlabel('x')
plt.ylabel('y')
#plt.title('Tri√°ngulo formado por los v√©rtices A, B y C con etiquetas')
plt.legend()
plt.show()
```
