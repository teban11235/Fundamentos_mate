```python
import matplotlib.pyplot as plt
import numpy as np

# Par√°metros para la primera elipse
a1, b1 = 5, 3  # Semiejes mayor y menor
c1 = np.sqrt(a1**2 - b1**2)  # Distancia focal
focos1 = [(-c1, 0), (c1, 0)]  # Focos de la elipse 1

# Par√°metros para la segunda elipse
a2, b2 = 5, 1.5  # Semiejes mayor y menor
c2 = np.sqrt(a2**2 - b2**2)  # Distancia focal
focos2 = [(-c2, 0), (c2, 0)]  # Focos de la elipse 2

# Crear valores de theta para graficar las elipses
theta = np.linspace(0, 2 * np.pi, 500)

# Coordenadas de la primera elipse
x1 = a1 * np.cos(theta)
y1 = b1 * np.sin(theta)

# Coordenadas de la segunda elipse
x2 = a2 * np.cos(theta)
y2 = b2 * np.sin(theta)

# Crear la gr√°fica
plt.figure(figsize=(10, 6))

# Graficar la primera elipse
plt.plot(x1, y1, label=f'Elipse 1: $a = {a1}, b = {b1}, e = {c1/a1:.2f}$', color='blue')
plt.scatter(*zip(*focos1), color='red', label='Focos Elipse 1', zorder=5)

# Graficar la segunda elipse
plt.plot(x2, y2, label=f'Elipse 2: $a = {a2}, b = {b2}, e = {c2/a2:.2f}$', color='green')
plt.scatter(*zip(*focos2), color='orange', label='Focos Elipse 2', zorder=5)

# Configuraci√≥n de la gr√°fica
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje X
plt.axvline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje Y
#plt.title('Dos elipses con distintas excentricidades')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')  # Escalas iguales para los ejes
plt.grid(alpha=0.4)
plt.legend()

# Mostrar la gr√°fica
plt.show()
```
