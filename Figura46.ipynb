```python
import matplotlib.pyplot as plt
import numpy as np

# Par√°metros de la par√°bola
foco = (0, 2)  # Coordenadas del foco
p = 2  # Distancia del v√©rtice al foco
directriz = -p  # Ecuaci√≥n de la directriz: y = -p

# Definir los valores de x para la par√°bola
x = np.linspace(-8, 8, 500)

# Ecuaci√≥n de la par√°bola: y = (x^2) / (4p)
y = (x**2) / (4 * p)

# Crear la gr√°fica
plt.figure(figsize=(8, 8))

# Graficar la par√°bola
plt.plot(x, y, label='Par√°bola: $y = \\frac{x^2}{4p}$', color='blue')

# Graficar el foco
plt.scatter(foco[0], foco[1], color='red', label=f'Foco {foco}', zorder=5)
plt.text(foco[0] + 0.3, foco[1], f'Foco {foco}', color='red', fontsize=10)

# Graficar la directriz
plt.axhline(directriz, color='green', linestyle='--', label=f'Directriz: $y = {directriz}$')

# Configuraci√≥n de la gr√°fica
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje X
plt.axvline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje Y
#plt.title('Par√°bola con v√©rtice en (0,0), foco en (0,2) y directriz $y = -2$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.4)
plt.legend()
plt.xlim(-8, 8)
plt.ylim(-3, 10)

# Mostrar la gr√°fica
plt.show()
```
