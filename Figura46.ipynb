```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la parábola
foco = (0, 2)  # Coordenadas del foco
p = 2  # Distancia del vértice al foco
directriz = -p  # Ecuación de la directriz: y = -p

# Definir los valores de x para la parábola
x = np.linspace(-8, 8, 500)

# Ecuación de la parábola: y = (x^2) / (4p)
y = (x**2) / (4 * p)

# Crear la gráfica
plt.figure(figsize=(8, 8))

# Graficar la parábola
plt.plot(x, y, label='Parábola: $y = \\frac{x^2}{4p}$', color='blue')

# Graficar el foco
plt.scatter(foco[0], foco[1], color='red', label=f'Foco {foco}', zorder=5)
plt.text(foco[0] + 0.3, foco[1], f'Foco {foco}', color='red', fontsize=10)

# Graficar la directriz
plt.axhline(directriz, color='green', linestyle='--', label=f'Directriz: $y = {directriz}$')

# Configuración de la gráfica
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje X
plt.axvline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)  # Eje Y
#plt.title('Parábola con vértice en (0,0), foco en (0,2) y directriz $y = -2$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.4)
plt.legend()
plt.xlim(-8, 8)
plt.ylim(-3, 10)

# Mostrar la gráfica
plt.show()
```
