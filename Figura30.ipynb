```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos para la funci√≥n y = 2x - 4
x = np.linspace(-1, 5, 500)  # M√°s puntos para una l√≠nea suave
y = 2 * x - 4

# Calcular intersecciones con los ejes
# Intersecci√≥n con el eje X (y = 0): 2x - 4 = 0 => x = 2
interseccion_x = np.array([2])
interseccion_y_x = np.array([0])

# Intersecci√≥n con el eje Y (x = 0): y = 2(0) - 4 => y = -4
interseccion_y = np.array([0])
interseccion_x_y = np.array([-4])

# Crear la gr√°fica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 2x - 4', color='blue')  # L√≠nea continua
plt.scatter(interseccion_x, interseccion_y_x, color='red', label='Intersecci√≥n con eje X', zorder=5)
plt.scatter(interseccion_y, interseccion_x_y, color='green', label='Intersecci√≥n con eje Y', zorder=5)

# Configuraci√≥n del gr√°fico
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)
plt.axvline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.6)
plt.xlabel('x')
plt.ylabel('y = 2x - 4')
#plt.title('Gr√°fica de la funci√≥n y = 2x - 4 con intersecciones')
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```
